```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.scatterplot(x=df.sepal_length,y=df.petal_length)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.sepal_length,df.petal_length,'o')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
```

```python
sns.displot(df['sepal_length'])
```

```python
df1 = sns.load_dataset('tips')
df1.head()
```

```python
###Q:Is there any relationship bet bill and tip
sns.scatterplot(x=df1.total_bill,y=df1.tip)
```

```python
sns.relplot(x=df1.total_bill,y=df1.tip)
```

```python
##Q: how may people are smoker and non smoker
df1['smoker'].value_counts()
```

```python
##Q: plot a graph smoker and non smoker 
sns.relplot(x=df1.total_bill,y=df1.tip,hue=df1.smoker) #relplot-realplot , hue-groupby &differentiate with colors
```

```python
sns.relplot(x=df1.total_bill,y=df1.tip,style=df1.smoker) #same operation like hue
```

```python
sns.relplot(x=df1.total_bill,y=df1.tip,style=df1.size)
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='time',hue='smoker') #groupby based on two columns
```

```python
##line graph
sns.relplot(x='total_bill',y='tip',data=df1,kind='line')
```

```python
#Two graphs 
sns.relplot(x='total_bill',y='tip',data=df1,col='time') #based on time column it separate two graph
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,col='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,col='sex')
```

```python
##categorical Plot
sns.catplot(x='day',y='total_bill',data=df1,)
```

```python
##Unnecessary plot we dont use this
sns.catplot(x='day',y='total_bill',data=df1,kind='swarm')
```

```python
sns.catplot(x='day',y='tip',data=df1,kind='box',hue='sex') #hue=day vs tip and sex
```
